```python
import pandas as pd
import numpy as np

  # Print current working directory
```

```python

with open('../data/msd_genre_dataset.txt', 'r') as file:
    lines = file.readlines()

cleaned_lines = []
for line in lines:
    if line.startswith('#'):
        continue 
    if line.startswith('%'):
        column_names = line[1:].strip().split(',')
        continue
    cleaned_lines.append(line.strip())

# Split the data into columns
data = [line.split(',') for line in cleaned_lines]
# Create a DataFrame from the data
df = pd.DataFrame(data, columns=column_names)
```

```python

df = df.replace('NULL',np.nan).dropna()
df.drop_duplicates(inplace=True)
```

```python
#import necessery objects from the sklearn library
from sklearn.preprocessing import StandardScaler,LabelEncoder 
scaler = StandardScaler()
label = LabelEncoder()
df[['loudness', 'tempo', 'duration']] = scaler.fit_transform(df[['loudness', 'tempo', 'duration']])
df['genre'] = label.fit_transform(df['genre'])
```

```python
#remove unecassery columns
drop_list = ["artist_name",'title','track_id']
df.drop(drop_list,axis = 1, inplace=True)
df.head()
```

```python
df.to_csv("../data/cleaned_msd_genre_dataSet.csv")
```
